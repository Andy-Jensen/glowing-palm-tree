```python
#Percentage of Pokemon whose Type 2 is Ghost
ghost_bois = len(pokemon[(pokemon['Type 1']== 'Ghost') | (pokemon['Type 2']== 'Ghost')])
round((ghost_bois / len(pokemon)) * 100, 2)
```

```python
pokemon[(pokemon['Type 1']== 'Ghost') | (pokemon['Type 2']== 'Ghost')]
```

```python
#Create a function named percent_ghost
def percent_ghost(df):
    ghost_bois = len(df[(df['Type 1']== 'Ghost') | (df['Type 2']== 'Ghost')])
    return round((ghost_bois / len(df)) * 100, 2)
```

```python
#Use the loaded dataframe as an argument
percent_ghost(pokemon)
```

```python
#Number of Pokemon whose Attack is greater than Defense
len(pokemon[pokemon['Attack'] > pokemon['Defense']])
```
