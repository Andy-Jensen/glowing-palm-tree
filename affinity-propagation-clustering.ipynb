# Get dummies and split the data

```python
#Get dummies for the data
dia=pd.get_dummies(columns=['cut', 'color', 'clarity'], data=dia)
```

```python
#Split the data
train, val, test=tts(dia)
```

```python
#data with dummies before scaling
train.head()
```

# Scaling

```python
#Create the scalar
mm=sklearn.preprocessing.MinMaxScaler()
```

```python
#set numeric columns to scale
numcol=['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
```

```python
#fit by train df's numeric columns
mm.fit(train[numcol])
```

```python
#reassign the numeric columns from the df to scaled versions
train[numcol]=mm.transform(train[numcol])
```

    <header>
    <h2>Visualizing the data</h2>
    </header>
    <dl>
        <dt>Clustering</dt>
        <dd>- use affinity propagation</dd>
        <dt>Affinity Propagation</dt>
        <dd>- clusters using exemplars</dd>
        <dd>- determines the best number of clusters on its own</dd>
        <dd>- sends messages between datapoints to determine the best exemplar for each cluster</dd>
    </dl>
</div>

```python
#reducing sample size to not crash the kernel
train_sample=train.sample(2000)
```

```python
#fitting the algorithm to train
af = AffinityPropagation(damping=.75, random_state=8675309)
af.fit(train_sample)

cluster_center_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_center_indices)
```

### Estimated number of clusters and silhouette score  
### silhouette score ranges from 1 to -1 closer to 1 is better and around 0 means there are overlapping clusters

```python
print(f'Estimated number of clusters: {n_clusters_}')
sils=metrics.silhouette_score(train_sample, labels, metric="sqeuclidean")
print(f'Silhouette Score: {sils}')
```

```python
#making predictions and adding it to a column in the sample df
train_sample['cluster']=af.fit_predict(train_sample)
```

```python
sns.scatterplot(x='cluster',y= 'price', data=train_sample)
```

```python
plt.close("all")
plt.figure(1)
plt.clf()

colors = plt.cycler("color", plt.cm.viridis(np.linspace(0, 1, 4)))

for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = train_sample[cluster_center_indices[k]]
    plt.scatter(
        train_sample[class_members, 0], train_sample[class_members, 1], color=col["color"], marker="."
    )
    plt.scatter(
        cluster_center[0], cluster_center[1], s=14, color=col["color"], marker="o"
    )
    for x in train_sample[class_members]:
        plt.plot(
            [cluster_center[0], x[0]], [cluster_center[1], x[1]], color=col["color"]
        )

plt.title("Estimated number of clusters: %d" % n_clusters_)
plt.show()
```
